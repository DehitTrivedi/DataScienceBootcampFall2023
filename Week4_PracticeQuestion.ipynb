Question 1

1. Create a line plot of ZN and INDUS in the housing data.

	a. For ZN, use a solid green line. For INDUS, use a blue dashed line.\

	b. Change the figure size to a width of 12 and height of 8.

	c. Change the style sheet to something you find https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html.


```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = 'boston_housing_data.csv'

data = pd.read_csv(file_path)

print(data.head())


# Set the style using a specific Matplotlib style sheet
plt.style.use('seaborn-darkgrid')  # Replace 'seaborn-darkgrid' with the desired style

# Set the figure size
plt.figure(figsize=(12, 8))

# Plot ZN with a solid green line
plt.plot(data['ZN'], color='green', linestyle='solid', label='ZN')

# Plot INDUS with a blue dashed line
plt.plot(data['INDUS'], color='blue', linestyle='--', label='INDUS')

# Set plot title and labels
plt.title('ZN and INDUS Plot')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

Question 2

2. Create a bar chart using col1 and col2 of dummy data.

	a. Give the plot a large title of your choosing.

	b. Move the legend to the lower-left corner.

	c. Do the same thing but with horizontal bars.
  
	d. Move the legend to the upper-right corner.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating dummy data
data = pd.DataFrame({
    'col1': [5, 10, 15, 20, 25],
    'col2': [3, 6, 9, 12, 15]
})

# Plotting a vertical bar chart
plt.figure(figsize=(8, 6))

# Plotting the bar chart
data.plot(kind='bar')
plt.title('Vertical Bar Chart - Title in the Center')
plt.legend(loc='lower left')
plt.show()
```

```python
# Plotting a horizontal bar chart
plt.figure(figsize=(8, 6))

# Plotting the horizontal bar chart
data.plot(kind='barh')
plt.title('Horizontal Bar Chart - Title in the Center')
plt.legend(loc='upper right')
plt.show()
```

Question 3

3. Create a histogram with pandas for using MEDV in the housing data.

	a. Set the bins to 20

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'boston_housing_data.csv'

data = pd.read_csv(file_path)

# Plotting a histogram for the 'MEDV' column
plt.figure(figsize=(8, 6))

# Creating the histogram with 20 bins
data['MEDV'].plot(kind='hist', bins=20)
plt.title('Histogram of MEDV')
plt.xlabel('MEDV Values')
plt.ylabel('Frequency')

plt.show()
```

4. Create a scatter plot of two heatmap entries that appear to have a very positive correlation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a sample dataset with two variables having a strong positive correlation
np.random.seed(42)
# Generating correlated data
x = np.random.normal(0, 1, 100)
y = x + np.random.normal(0, 0.5, 100)

# Creating a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Plotting a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Two Variables with Positive Correlation')
plt.show()
```

Question 5

5. Now, create a scatter plot of two heatmap entries that appear to have negative correlation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generating a sample dataset with two variables having a negative correlation
np.random.seed(42)
# Generating negatively correlated data
x = np.random.normal(0, 1, 100)
y = -x + np.random.normal(0, 0.5, 100)

# Creating a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Plotting a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data['X'], data['Y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Two Variables with Negative Correlation')
plt.show()
```
